```python
def call_result_count(result):
  li = [0,0]
  for i in result.itertuples():
    if i[1] == 'CEVAP VERMED ' or i[1] =='CEVAP VERMEDİ':
      li[0]+=1
    else: li[1]+=1
  return li
```

```python

def methods_count(method):
  li = [0,0,0,0,0]
  for i in method.itertuples():
    if i[1] == 'ARAMA':
      li[0]+=1
    elif i[1] == 'E-POSTA':
      li[1]+=1
    elif i[1] == 'E-POSTA':
      li[2]+=1
    elif i[1] == 'SOSYAL MEDYA':
      li[3]+=1
    else:
      li[4]+=1
  return li
```

```python
def todays_results(df):
  # modifiy columns
  df.columns = df.loc[0,].apply(lambda x: str(x).strip())

  # rewrite_dates
  df["date_1"] = rewrite_date(df['1. İRTİBAT TARİHİ'])
  df["date_2"] = rewrite_date(df['2. İRTİBAT TARİHİ'])
  df["date_3"] = rewrite_date(df['3. İRTİBAT TARİHİ'])
  df["date_4"] = rewrite_date(df['4. İRTİBAT TARİHİ'])

  # Today?
  today = datetime.now().strftime("%Y-%m-%d")

  # calls Series
  calls_series = pd.Series( [df["date_1"], df["date_2"], df["date_3"], df["date_4"]])
  calls_nums = (calls_series.apply(lambda x : (x==today).sum()))
  
  # filtering the method of contact
  
  method_1 = pd.DataFrame(df['1. İRTİBAT ŞEKLİ'].loc[(df["date_1"]==today)])
  method_2 = pd.DataFrame(df['2. İRTİBAT ŞEKLİ'].loc[(df["date_2"]==today)])
  method_3 = pd.DataFrame(df['3. İRTİBAT ŞEKLİ'].loc[(df["date_3"]==today)])
  method_4 = pd.DataFrame(df['4. İRTİBAT ŞEKLİ'].loc[(df["date_4"]==today)])


 # total methods
  methods_count_list = pd.Series([method_1,method_2,method_3,method_4]).apply(lambda x: np.array(methods_count(x))).sum()

# filtering results
  result_1 = pd.DataFrame(df['1. İRTİBAT ŞEKLİ'].loc[(df["date_1"]==today)])
  result_2 = pd.DataFrame(df['2. İRTİBAT ŞEKLİ'].loc[(df["date_2"]==today)])
  result_3 = pd.DataFrame(df['3. İRTİBAT ŞEKLİ'].loc[(df["date_3"]==today)])
  result_4 = pd.DataFrame(df['4. İRTİBAT ŞEKLİ'].loc[(df["date_4"]==today)])

 # total results
  results_count_list = pd.Series([result_1,result_2,result_3,result_4]).apply(lambda x: np.array(call_result_count(x))).sum()
  print(results_count_list)
  # The Results:
  print(
  f"""
  Bugün Özeti:
  Toplam Firma : {calls_nums.sum()}
  ---------------------------------
  Arama : {methods_count_list[0]}
  WhatsApp : {methods_count_list[1]}
  Email : {methods_count_list[2]}
  Sosyal Medya : {methods_count_list[3]}
  NA : {methods_count_list[4]}
  ---------------------------------
  1. İRTİBAT : {calls_nums[0]}
  2. İRTİBAT : {calls_nums[1]}
  3. İRTİBAT : {calls_nums[2]}
  4. İRTİBAT : {calls_nums[3]}
  _________________________________
  \n{method_1.value_counts() if not str(method_1).startswith("E") else "1. İRTİBAT YOK"}
  _________________________________
  \n{method_2.value_counts() if not str(method_2).startswith("E") else "2. İRTİBAT YOK"}
  _________________________________
  \n{method_3.value_counts() if not str(method_3).startswith("E") else "3. İRTİBAT YOK"}
  _________________________________
  \n{method_4.value_counts() if not str(method_4).startswith("E") else "4. İRTİBAT YOK"}

  """)

  


                                                               

```
